```python
# =========================
# TASK 1: IRIS DATASET ANALYSIS
# =========================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Styling
sns.set(style="whitegrid")

# =========================
# Load Dataset
# =========================
df = sns.load_dataset("iris")

# =========================
# Dataset Exploration
# =========================
print("Shape of dataset:", df.shape)
print("\nColumns:", df.columns)
print("\nFirst 5 rows:\n", df.head())
print("\nInfo:\n")
df.info()
print("\nDescription:\n", df.describe())

# =========================
#  Scatter Plot 1
# Sepal Length vs Sepal Width
# =========================
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()

# =========================
# Scatter Plot 2
# Petal Length vs Petal Width
# =========================
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
plt.title("Petal Length vs Petal Width")
plt.show()

# =========================
# Histogram (Distributions)
# =========================
df.hist(figsize=(10,8), bins=20)
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()

# =========================
# Box Plot (Outliers Check)
# =========================
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Box Plot of Iris Features")
plt.show()

# =========================
# 🧠 Conclusion (Print Form)
# =========================
print("""
CONCLUSION:

The Iris dataset contains 3 species of flowers with 4 numerical features.
Petal length and petal width show strong separation between species,
making them very useful for classification.

The dataset has no major missing values or extreme outliers.
Visualizations helped understand patterns and relationships clearly.
""")
```
